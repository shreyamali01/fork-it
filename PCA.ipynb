```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import os
```

```python
train_data = '/Users/shreya/Documents/fork-it/HAR/Combined/Train/LAYING/Subject_1.csv'
df = pd.read_csv(train_data)

print(df)
```

```python
df['total_acc'] = np.sqrt(df['accx']**2+ df['accy']**2+ df['accz']**2)
print(df)
```

```python
pca = PCA(n_components=2)
pca_features = pca.fit_transform(df[['accx','accy','accz']])

plt.scatter(pca_features[:, 0], pca_features[:, 1])
plt.title("PCA on Total Acceleration - LAYING - Subject 1")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.show()
```
